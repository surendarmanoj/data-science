This is the code section to implement Simple Linear Regression.

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Genearting the necessary data. We have used 500 points for the refernce

data = []

for i in range(500):
  cgpa = round(random.uniform(4, 10), 2)
  pkg = 0.8 * cgpa - 2

  package = round(pkg + random.uniform(-.5, .5), 2)

  package = max(package, 1)

  data.append([cgpa, package])

df = pd.DataFrame(data, columns=['cgpa','package'])
print(df.head())
```

```python
df.shape
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
# initiating the linear regression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
m = lr.coef_
```

```python
c = lr.intercept_
```

```python
plt.scatter(df['cgpa'], df['package'])
#  drawing the best fit line
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
# Testing our output with the best fit line equation

# 0  9.93     5.50
# 1  7.22     3.62
# 2  7.38     4.34
# 3  7.31     3.53
# 4  7.90     4.38

# y = mx+c
op = []

x_ip = [8.5, 9.5, 100]
# x_ip = [9.93, 7.22, 7.38, 7.31, 7.9]
for j in x_ip:
  op.append( (m * j) + c)

op
```
