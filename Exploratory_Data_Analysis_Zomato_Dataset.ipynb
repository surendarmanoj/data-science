**Rationale for Dropping Features:**

The original dataset included additional features that were not essential for our analysis, such as:

**URLs** ('photos_url', 'url', 'deeplink', 'menu_url', 'book_url', 'featured_image', 'events_url', 'order_deeplink', 'order_url')

**API keys and IDs** ('apikey', 'id', 'R.res_id')

**Location coordinates** ('location.latitude', 'location.longitude', 'location.city_id', 'location.zipcode', 'location.locality')

**Redundant or less informative fields** ('thumb', 'user_rating.rating_color')
By excluding these features, we focused on the most relevant data, simplifying the dataset for more effective analysis.
