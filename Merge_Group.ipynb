```python
import pandas as pd
```

```python
df = pd.read_excel('Group by - Merge.xlsx','overall')
df
```

```python
df1 = df.copy()
```

```python
df2 = df1.groupby('City')
df2
```

```python
for city, g_df in df2:
  print(city)
  print(g_df)
```

```python
df_c = pd.read_excel('Group by - Merge.xlsx','Chennai')
df_c
```

```python
df_h = pd.read_excel('Group by - Merge.xlsx','Hydrabad')
df_h
```

```python
df_b = pd.read_excel('Group by - Merge.xlsx','Bangalore')
df_b
```

```python
df5 = pd.concat([df_c, df_b], keys=['Chennai','Bangalore'])
df5
```

```python
df5.loc['Chennai']
```

```python
df6 = pd.concat([df5, df_h])
df6
```

```python
w_df = {
    'city':['Chennai','Bangalore','Kochi'],
    'windSpeed':[7,4,8]
}
w_df
df_w = pd.DataFrame(w_df)
df_w
```

```python
t_df = {
    'city':['Bangalore','Hydrabad','Goa'],
    'temperature':[32, 34,33]
}
t_df
df_t= pd.DataFrame(t_df)
df_t
```

```python
df7 = pd.concat([df_t, df_w], axis = 1)
df7
```

```python
s = pd.Series(['Humid','Dry','Rain'], name='Event')
s
```

```python
df8 = pd.concat([df7, s], axis=1)
df8
```

```python
pd.merge(df_w, df_t, on='city', how='outer', indicator=True)
```
