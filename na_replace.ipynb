```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("missing.csv", parse_dates=['Date'])
df
```

```python
df1 = df.copy()
type(df1.Date[0])
```

```python
df1.set_index('Date')
df1
```

```python
df2 = df1.set_index('Date')
df2
```

```python
df1.set_index('Date', inplace=True)
df1
```

```python
df3 = df2.fillna(0)
df3
```

```python
df3 = df2.fillna({
    'Temperature' : 0,
    'Windspeed' : 0,
    'Event' : 'Nothing'
})
df3
```

```python
df3 = df2.fillna(method='ffill')
df3
```

```python
df3 = df2.fillna(method='bfill')
df3
```

```python
df3 = df2.fillna(method='ffill', limit=2)
df3
```

```python
df3 = df2.interpolate(method='time')
df3
```

```python
dt = pd.date_range('2024-01-01','2024-01-15')
idx = pd.DatetimeIndex(dt)
df3 = df2.reindex(idx)
df3
```

```python
df3 = df2.dropna(how='all')
df3
```

```python
df3 = df2.dropna(thresh=2)
df3
```

```python
df4 = pd.read_csv('replace.csv')
df4
```

```python
df5 = df4.replace([-9999,32], np.NaN)
df5
```

```python
df5 = df4.replace({
    'Temperature':-9999,
    'Windspeed' : 7
}, np.NaN)
df5
```

```python
df5 = df4.replace({
    -9999:np.NaN,
    'no event':'yes'
})
df5
```
