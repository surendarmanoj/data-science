```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data for height (in cm) and weight (in kg)
heights = np.random.normal(loc=170, scale=10, size=1000)  # Mean height = 170 cm, Std dev = 10
weights = np.random.normal(loc=70, scale=15, size=1000)   # Mean weight = 70 kg, Std dev = 15

# Create a DataFrame
data = pd.DataFrame({
    'Height': heights,
    'Weight': weights
})

data.head()
```

```python
# Function for min-max normalization
def min_max_normalization(df):
    return (df - df.min()) / (df.max() - df.min())
```

```python
normalized_data = min_max_normalization(data)
```

```python
# Save the data to a CSV file for review or further use
data.to_csv('synthetic_height_weight.csv', index=False)
normalized_data.to_csv('normalized_height_weight.csv', index=False)
```

```python
# Display a preview of the normalized data
print(normalized_data.head())
```

```python
data.head()
```

```python

# Calculate means
mean_height_before = data['Height'].mean()
mean_weight_before = data['Weight'].mean()
mean_height_after = normalized_data['Height'].mean()
mean_weight_after = normalized_data['Weight'].mean()

# Scatter plot with means marked
plt.figure(figsize=(12, 6))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(data['Height'], data['Weight'], alpha=0.5, label="Data points")
plt.scatter(mean_height_before, mean_weight_before, color='red', label="Mean", s=100)
plt.title('Before Scaling')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(True)

# Normalized data
plt.subplot(1, 2, 2)
plt.scatter(normalized_data['Height'], normalized_data['Weight'], alpha=0.5, label="Data points")
plt.scatter(mean_height_after, mean_weight_after, color='red', label="Mean", s=100)
plt.title('After Scaling')
plt.xlabel('Height (Normalized)')
plt.ylabel('Weight (Normalized)')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

plt.show()
```
