```python
sns.countplot(y = 'Product line',hue='Gender', data=df1, palette='hsv_r')
```

```python
sns.countplot(y = 'Product line', data=df1, facecolor=(0,1,0,1), linewidth=4, edgecolor=sns.color_palette("dark",1))
```

```python
sns.barplot(y='Product line', x='Total', hue='Gender', data=df1, order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=['Male','Female'], ci=None, palette='Dark2', estimator=np.median)
```

```python
x = df1['Product line'].sort_values()
x.unique()
```

```python
df1.columns
```

```python
sns.boxplot(y='Total',x = 'City',hue='Gender', data=df1, width=0.25, hue_order=['Male', 'Female'], showmeans=True, meanprops={
    "marker":"o",
    "markerfacecolor":"white",
    "markersize":10
}, palette="CMRmap", linewidth=5)
```

```python
sns.boxplot(data=df1)
```

```python
sns.violinplot(y='Total', data=df1)
```

```python
sns.violinplot(y='Unit price', data=df1)
```

```python
sns.violinplot(y='Total',x ="Payment", hue="Gender", data=df1, split=True, inner='stick', cut=0)
```

```python
sns.stripplot(y='Total',x='Payment', data=df1, jitter=.1)
sns.boxplot(y='Total',x='Payment', data=df1, color='black')
```

```python
sns.swarmplot(data=df1,y='Total', x='Payment')
sns.violinplot(data=df1,y='Total', x='Payment')
```

```python
df1.columns
```

```python
sns.catplot(data=df1, x='Payment', y='Total', kind='violin', col='Gender', row='Customer type', palette='gray')
```

```python
sns.histplot(data=df1, x='Total', stat='probability', kde=True)
```

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(x='Unit price',y='gross income',hue='Gender', data=df1)
sns.rugplot(x='Unit price',y='gross income',hue='Gender', data=df1, height=-.05, clip_on=False)
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(x='Unit price', data=df1)
```

```python
# ECDF - Emperical Cumulative Distribution Function
sns.ecdfplot(y='gross income',hue='Gender', data=df1, stat='count')
```

```python
sns.displot(x='Total', data=df1, kde=True, rug=True, rug_kws={'height':.05}, kde_kws={'bw_adjust':5}, hue='Gender', multiple='stack', col="Payment")
```
